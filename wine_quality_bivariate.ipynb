<h2 style="color: #008080;">Continuous Vs Continuous</h2>

```python
# Calculating the Pearson correlation coefficient between 'SalePrice' and 'GrLivArea' columns
wine3['quality'].corr(wine3['citric_acid']) # PEARSON
```

```python
numerical_columns_include = wine3.select_dtypes(include='number').columns  # Seleccionar las columnas numéricas

correlation_results = {}

for i in range(len(numerical_columns_include)):
    for j in range(i+1, len(numerical_columns_include)):  # Evitar duplicados, como calcular (x, y) y (y, x)
        col1 = numerical_columns_include[i]
        col2 = numerical_columns_include[j]
        
        # Calcular la correlación de Pearson
        corr_value = wine3[col1].corr(wine3[col2])
        
        # Guardar el resultado
        correlation_results[f'{col1} vs {col2}'] = corr_value
```

```python
### Spearman Correlation 
```

```python
from scipy.stats import pearsonr, spearmanr

# Seleccionar las columnas numéricas
numerical_columns_include = wine3.select_dtypes(include='number').columns

# Iterar sobre todas las combinaciones de columnas numéricas
for i in range(len(numerical_columns_include)):
    for j in range(i+1, len(numerical_columns_include)):  # Para evitar duplicados, como (x, y) y (y, x)
        col1 = numerical_columns_include[i]
        col2 = numerical_columns_include[j]
        
        # Calcular la correlación de Pearson
        pearson_corr, _ = pearsonr(wine3[col1], wine3[col2])
        
        # Calcular la correlación de Spearman
        spearman_corr, _ = spearmanr(wine3[col1], wine3[col2])
        
        # Imprimir los resultados juntos
        print(f"Correlación de Pearson y Spearman para {col1} vs {col2}:")
        print(f"  Pearson: {pearson_corr:.2f}, Spearman: {spearman_corr:.2f}")
        print()
```

```python
num_corr = wine3.corr().round(2)
```

```python
mask = np.zeros_like(num_corr)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(25, 15))
sns.set(font_scale=1.5) 

ax = sns.heatmap(num_corr, mask=mask, annot=True, annot_kws={"size": 12}, linewidths=.5, cmap="coolwarm", fmt=".2f", ax=ax) # round to 2 decimal places
ax.set_title("Dealing with Multicollinearity", fontsize=20) 
plt.show()
```
